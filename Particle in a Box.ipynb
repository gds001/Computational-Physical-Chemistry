```python
import numpy as np
import matplotlib.pyplot as plt
import math as math
import random as random
```

```python
def psi1DPIB(n,x,L):
    return math.sqrt(2/L)*math.sin(n*math.pi*x/L)

def energy1DPIB(n,L):
    return math.pi*math.pi*n*n/(2*m*L*L)

def integrate(x,y):
    result=0
    for i in range(1,L*numPoints,1):
        temp=(y[i]+y[i-1])/2*(x[i]-x[i-1])
        result+=temp
    return result

def psi2DPIB(nx,ny,x,y,Lx,Ly):
    return psi1DPIB(nx,x,Lx)*psi1DPIB(ny,y,Ly)

def energy2DPIB(nx,ny,Lx,Ly):
    return energy1DPIB(nx,Lx)+energy1DPIB(ny,Ly)
```

```python
m=1 #me
L=5 #ao
numPoints=1000

x=np.array([])
psi1DPIB1=np.array([])
psi1DPIB2=np.array([])

for i in range(L*numPoints):
    x=np.append(x,[float(i/numPoints)])
    psi1DPIB1=np.append(psi1DPIB1,[psi1DPIB(1,float(i/numPoints),L)])
    psi1DPIB2=np.append(psi1DPIB2,[psi1DPIB(2,float(i/numPoints),L)])

plt.plot(x,psi1DPIB1)
plt.plot(x,psi1DPIB2)
```

```python
pos1DPIB1=np.array([])
pos1DPIB2=np.array([])
pos1DPIB12=np.array([])

for i in range(L*numPoints):
    pos1DPIB1=np.append(pos1DPIB1,[psi1DPIB1[i]*psi1DPIB1[i]])
    pos1DPIB2=np.append(pos1DPIB2,[psi1DPIB2[i]*psi1DPIB2[i]])
    pos1DPIB12=np.append(pos1DPIB12,[psi1DPIB1[i]*psi1DPIB2[i]])

plt.plot(x,pos1DPIB1)
plt.plot(x,pos1DPIB2)

print("{0:.2f}".format(integrate(x,pos1DPIB1)))
print("{0:.2f}".format(integrate(x,pos1DPIB2)))
print("{0:.2f}".format(integrate(x,pos1DPIB12)))
```

```python
plt.plot(x,x)
plt.plot(x,psi1DPIB1)

psixpsi1=np.array([])

for i in range(L*numPoints):
    psixpsi1=np.append(psixpsi1,[psi1DPIB1[i]*x[i]*psi1DPIB1[i]])

print("{0:.2f}".format(integrate(x,psixpsi1)))
```

```python
psi12=np.array([])
pos12=np.array([])
psixpsi12=np.array([])

for i in range(L*numPoints):
    psi12=np.append(psi12,[(psi1DPIB1[i]+psi1DPIB2[i])/math.sqrt(2)])
    pos12=np.append(pos12,[psi12[i]*psi12[i]])
    psixpsi12=np.append(psixpsi12,pos12[i]*x[i])

plt.plot(x,psi12)
plt.plot(x,pos12)
plt.plot(x,psixpsi12)

print("{0:.2f}".format(integrate(x,pos12)))
print("{0:.2f}".format(integrate(x,psixpsi12)))
```
